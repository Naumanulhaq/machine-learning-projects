```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# load titanic data from the seaborn database
titanic_data = sns.load_dataset('titanic')
titanic_data.head()
```

```python
titanic_data.describe()
```

```python
# check for null values
titanic_data.isnull().sum()
```

```python
titanic_data.head()
```

```python
# fill null values with the median value in age column
titanic_data['age'].fillna(titanic_data['age'].median(), inplace=True)
# drop
titanic_data.dropna(subset=['embarked'], inplace=True)
# convert 'sex' to numeric from string (male:0, female:1)
titanic_data['sex'].replace({'male':1, 'female':0}, inplace=True)
titanic_data['alone'].replace({True:1, False:0}, inplace=True)
titanic_data['adult_male'].replace({True:1, False:0}, inplace=True)
titanic_data.head()
```

```python
x = titanic_data[['sex', 'age', 'sibsp', 'parch', 'alone', 'adult_male']]
y = titanic_data['survived']
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=2)
model = LogisticRegression()
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
accuracy_score(Y_test, y_pred)
```

```python
model.coef_
```

```python
import joblib
joblib.dump(model, 'titanic_logistic_model.pkl')
print("Model saved")
```
