```python
# import the libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pandas as pd
```

```python
# read the csv file
df=pd.read_csv('/content/salary_data_cleaned.csv')
df.head(1)
```

```python
# get the column names with numbers
df_columns=df.select_dtypes(include=['number']).columns
df_columns
# only get these columns from the dataframe
df=df[df_columns]
df.head(5)
# split the dataset into 70:30 ratio
from sklearn.model_selection import train_test_split
train_df,test_df=train_test_split(df,test_size=0.3,random_state=42)
# check the training and validation number of records
train_df.shape, test_df.shape
```

```python
# preparing the training data to train the model on learning the avg_salary
train_x=train_df.drop(['avg_salary'],axis=1)
train_y=train_df['avg_salary']
```

```python
# train the model on the training dataset
linReg=LinearRegression()
linReg.fit(train_x,train_y)
```

```python
test_df.head()
```

```python
# prepare the test data to predict the model efficiency
test_x=test_df.drop(['avg_salary'],axis=1)
test_y=test_df['avg_salary']
```

```python
# see the mean square error to check how good the model has performed in learning
pred_y=linReg.predict(test_x)
mean_squared_error(test_y,pred_y)
```
